# Exploratory Data Analysis

# Gender

```python
df.columns
```

```python
ax=sns.countplot(x="Gender",data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales=df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sns.barplot(x="Gender",y="Amount",data=sales)
```

1.From the above graph we can see that women are the majority buyers.

2.Also the amount spent by women are more than that of men

# Age

```python
ax=sns.countplot(data=df, x="Age Group", hue="Gender")
# hue provides a divison between the males and females
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Totalamount vs Age Group
sales_age=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sales_age
```

```python
sns.barplot(x="Age Group", y="Amount", data=sales_age)
```

 From the above graph we can see that most of the buyers are of the age group between 26-35 years female

# State

```python
# Total number of orders from 10 states
sales_states=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sales_states
```

```python
# Giving size to the chart
sns.set(rc={"figure.figsize":(15,8)})
sns.barplot(data=sales_states, x="State", y="Orders")
```

```python
# Total amount from 10 states
sales_states=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sales_states
```

```python
sns.set(rc={"figure.figsize":(15,5)})
sns.barplot(data=sales_states, x="State", y="Amount")
```

 From the above graphs we can see that most orders are from up,Maharashtrs and karnataka. Also these are the top three states to invest money on the sales.

# Marital Status

```python
ax=sns.countplot(data=df, x="Marital_Status")
sns.set(rc={"figure.figsize":(5,4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_married=df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sales_married
```

```python
sns.set(rc={"figure.figsize":(15,15)})
sns.barplot(data=sales_married, x="Marital_Status", y="Amount",hue="Gender")
```

 From the above graphs we can see that most of the buyers are married (women)and they have high purchasing power


# Occupation

```python
sns.set(rc={"figure.figsize":(15,15)})
ax=sns.countplot(data=df, x="Occupation")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occu=df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sales_occu
```

```python
sns.set(rc={"figure.figsize":(15,15)})
sns.barplot(data=sales_occu, x="Occupation", y="Amount",)
```

From the above graphs we can see that the most buyers are from IT, Aviation and Healthcare Sector. 

# Product Category

```python
sns.set(rc={"figure.figsize":(15,15)})
ax=sns.countplot(data=df, x="Product_Category")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_cat=df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sales_cat
```

```python
sns.barplot(data=sales_cat, x="Product_Category", y="Amount",)
sns.set(rc={"figure.figsize":(20,5)})
```

From the above graph we can see that most of the sold products are from food, Footwear and Electronics category
